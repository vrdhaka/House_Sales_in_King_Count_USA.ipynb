```python
# ============================
# 0) SETUP — imports & style
# ============================
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression, Ridge
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.pipeline import Pipeline
from sklearn.metrics import r2_score

plt.rcParams["figure.figsize"] = (8, 5)
plt.rcParams["figure.dpi"] = 120

print("Libraries imported.")
```

```python
# ===========================================
# 1) LOAD DATA — multiple mirrors + fallback
# ===========================================
# If you already uploaded a CSV via the left sidebar (Upload button),
# set local_name to that filename (e.g., "kc_house_data.csv")
local_name_candidates = ["kc_house_data.csv", "kc_house_data_NaN.csv"]

# Public mirrors (any one may work in your browser):
url_candidates = [
    # GitHub Gist (large raw CSV)
    "https://gist.githubusercontent.com/guilhermesilveira/eb03c2f0d0fc52d4df2528b330f50914/raw/kc_house_data.csv",
    # GitHub raw (practical_statistics repo)
    "https://raw.githubusercontent.com/nchelaru/practical_statistics/master/kc_house_data.csv",
    # Another GitHub raw mirror
    "https://raw.githubusercontent.com/basilhan/datasets/master/kc-house-data.csv",
]

df = None

# Try local files first (if you uploaded)
for name in local_name_candidates:
    try:
        df = pd.read_csv(name)
        print(f"Loaded local file: {name}")
        break
    except Exception:
        pass

# Try URLs next
if df is None:
    for url in url_candidates:
        try:
            df = pd.read_csv(url)
            print(f"Loaded from URL: {url}")
            break
        except Exception:
            pass

# If still None, create a synthetic fallback with the same key columns
if df is None:
    print("⚠️ Could not load remote/local CSV. Using synthetic fallback data so you can finish the project.")
    rng = np.random.default_rng(42)
    n = 5000

    # Create realistic-ish features
    bedrooms = rng.integers(1, 7, size=n)
    bathrooms = rng.uniform(1, 4.5, size=n).round(1)
    floors = rng.integers(1, 4, size=n)
    waterfront = rng.integers(0, 2, size=n)  # 0/1
    view = rng.integers(0, 5, size=n)        # 0..4
    condition = rng.integers(1, 6, size=n)   # 1..5
    grade = rng.integers(1, 14, size=n)      # 1..13
    sqft_above = rng.normal(1800, 600, size=n).clip(400, 7000).astype(int)
    sqft_basement = rng.integers(0, 1500, size=n)
    sqft_living = (sqft_above + 0.7 * sqft_basement).astype(int)
    sqft_living15 = (sqft_living * rng.uniform(0.9, 1.1, size=n)).astype(int)
    lat = rng.uniform(47.1, 47.8, size=n)
    longi = rng.uniform(-122.5, -121.3, size=n)

    # Price model with noise
    price = (
        80000
        + sqft_living * 320
        + bedrooms * 8000
        + bathrooms * 12000
        + floors * 5000
        + waterfront * 200000
        + view * 15000
        + grade * 25000
        + (lat - 47.5) * 200000
        + rng.normal(0, 60000, size=n)
    ).clip(50000, 3000000)

    df = pd.DataFrame({
        "price": price.astype(float),
        "bedrooms": bedrooms,
        "bathrooms": bathrooms,
        "sqft_living": sqft_living,
        "sqft_lot": rng.integers(1000, 50000, size=n),
        "floors": floors,
        "waterfront": waterfront,
        "view": view,
        "condition": condition,
        "grade": grade,
        "sqft_above": sqft_above,
        "sqft_basement": sqft_basement,
        "yr_built": rng.integers(1900, 2015, size=n),
        "yr_renovated": rng.choice([0, 1990, 2005, 2010, 2014], size=n, p=[0.7, 0.1, 0.08, 0.07, 0.05]),
        "zipcode": rng.choice([98001,98002,98003,98004,98005,98006,98007,98008,98010,98198,98199], size=n),
        "lat": lat,
        "long": longi,
        "sqft_living15": sqft_living15,
        "sqft_lot15": rng.integers(1000, 50000, size=n),
    })
    print("✅ Synthetic fallback dataset created.")

print("Rows, Columns:", df.shape)
df.head()
```

```python
# ===========================================================
# Q2) Drop 'id' and 'Unnamed: 0' then describe() (screenshot)
# ===========================================================
df.drop(["id", "Unnamed: 0"], axis=1, inplace=True, errors='ignore')
df.describe(include='all')
```

```python
# ==========================================================
# Q4) Boxplot (matplotlib): price vs waterfront (screenshot)
# ==========================================================
df.boxplot(column='price', by='waterfront', grid=False)
plt.title("Price distribution by Waterfront (0 = No, 1 = Yes)")
plt.suptitle("")  # remove duplicated super-title from pandas
plt.xlabel("Waterfront")
plt.ylabel("Price")
plt.show()
```

```python
# ==========================================================
# Q5) Scatter + regression line: sqft_above vs price (shot)
# ==========================================================
x = df['sqft_above'].to_numpy()
y = df['price'].to_numpy()

plt.scatter(x, y, alpha=0.3)
m, b = np.polyfit(x, y, 1)
plt.plot(x, m*x + b)
plt.title("sqft_above vs price (with regression line)")
plt.xlabel("sqft_above")
plt.ylabel("price")
plt.show()
```

```python
# ====================================
# Train/Test split for Q6..Q10 utility
# ====================================
target = 'price'
y_all = df[target]

# Q6 single feature
X_sqft = df[['sqft_living']].copy()

# Q7..Q10 multi-feature set
features = [
    "floors",
    "waterfront",
    "lat",
    "bedrooms",
    "sqft_basement",
    "view",
    "bathrooms",
    "sqft_living15",
    "sqft_above",
    "grade",
    "sqft_living"
]

model_df = df[features + [target]].dropna()
X_multi = model_df[features].copy()
y_multi = model_df[target].copy()

X_sqft_train, X_sqft_test, y_sqft_train, y_sqft_test = train_test_split(
    X_sqft, y_all, test_size=0.15, random_state=1
)
X_train, X_test, y_train, y_test = train_test_split(
    X_multi, y_multi, test_size=0.15, random_state=1
)

print("Single-feature shapes:", X_sqft_train.shape, X_sqft_test.shape)
print("Multi-feature  shapes:", X_train.shape, X_test.shape)
```

```python
# ============================================
# Q6) LinearRegression on sqft_living (shot)
# ============================================
lin1 = LinearRegression()
lin1.fit(X_sqft_train, y_sqft_train)

y_sqft_pred = lin1.predict(X_sqft_test)
r2_q6 = r2_score(y_sqft_test, y_sqft_pred)

print("Intercept:", lin1.intercept_)
print("Coefficient for sqft_living:", lin1.coef_[0])
print("R^2 on test:", r2_q6)
```

```python
# ===================================================
# Q7) LinearRegression on the 11 features list (shot)
# ===================================================
lin_multi = LinearRegression()
lin_multi.fit(X_train, y_train)

y_pred_lin_multi = lin_multi.predict(X_test)
r2_q7 = r2_score(y_test, y_pred_lin_multi)

print("Coefficients:\n", pd.Series(lin_multi.coef_, index=X_train.columns))
print("Intercept:", lin_multi.intercept_)
print("R^2 on test:", r2_q7)
```

```python
# =====================================================================
# Q8) Pipeline: Scale -> Polynomial(deg=2) -> LinearRegression (shot)
# =====================================================================
pipe = Pipeline(steps=[
    ("scale", StandardScaler(with_mean=False)),
    ("poly", PolynomialFeatures(degree=2, include_bias=False)),
    ("linreg", LinearRegression())
])

pipe.fit(X_train, y_train)
y_pred_pipe = pipe.predict(X_test)
r2_q8 = r2_score(y_test, y_pred_pipe)

print("R^2 on test (Pipeline degree=2):", r2_q8)
```

```python
# ==============================================
# Q9) Ridge(alpha=0.1) on original features (shot)
# ==============================================
ridge_pipe = Pipeline(steps=[
    ("scale", StandardScaler()),
    ("ridge", Ridge(alpha=0.1, random_state=1))
])

ridge_pipe.fit(X_train, y_train)
y_pred_ridge = ridge_pipe.predict(X_test)
r2_q9 = r2_score(y_test, y_pred_ridge)

print("R^2 on test (Ridge alpha=0.1):", r2_q9)
```

```python
# ==========================================================
# Q10) Poly(deg=2) -> Ridge(alpha=0.1) on poly features (shot)
# ==========================================================
poly = PolynomialFeatures(degree=2, include_bias=False)
X_train_poly = poly.fit_transform(X_train)
X_test_poly  = poly.transform(X_test)

ridge_poly_pipe = Pipeline(steps=[
    ("scale", StandardScaler(with_mean=False)),
    ("ridge", Ridge(alpha=0.1, random_state=1))
])

ridge_poly_pipe.fit(X_train_poly, y_train)
y_pred_ridge_poly = ridge_poly_pipe.predict(X_test_poly)
r2_q10 = r2_score(y_test, y_pred_ridge_poly)

print("R^2 on test (Poly degree=2 + Ridge alpha=0.1):", r2_q10)
```
